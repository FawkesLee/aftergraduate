# `torch.nn.init`中实现的初始化函数
[torch.init中的各函数](https://pytorch.org/docs/stable/nn.html#torch-nn-init)
## 均匀分布
`torch.nn.init.uniform_(tensor, a=0, b=1)`~U(a,b)
## 正态分布
`torch.nn.init.normal_(tensor, mean=0, std=1)`~N(mean,std)，默认为0,1
## 初始化为常数
`torch.nn.init.constant_(tensor, val)`

初始化整个矩阵为常数`val`
## Xavier
基本思想是通过网络层时，**输入和输出方差相同**，包括前向传播和反向传播
### 起因
以Sigmoid为例:
* 若初始化值很小，那么随着层数的传递，方差会趋于0，此时输入值也变得越来越小，在Sigmoid的零点附近，接近于线性，失去非线性。
* 如果初始值很大，那么随着层数的传递，方差会迅速增加，此时输入值变得很大，而sigmoid在大输入值下导数趋近于0，反向传播时会遇到梯度消失的问题
若期望神经网络正常运行，则是用正确的权重初始化网络非常重要。在开始训练网络之前，需要确保权重在合理范围内，此即Xavier初始化被提出的起因：**在每一层网络保证输入和输出的方差相同**。
### 实现方式
Xavier初始化的实现就是下面的均匀分布：![Xavier][base64str1]
对于Xavier初始化方式，pytorch提供两种:uniform和normal
* `torch.nn.init.xavier_uniform_(tensor, gain=1)` 均匀分布~U(-a,a)
    其中，a的计算公式:![a][base64str2]
* `torch.nn.init.xavier_normal(tensor,gain=1)` 正态分布~N(0,std),其中,std的计算公式：![b][base64str3]

**fain_in**:是指第i层神经元个数

**fain_out**:是指第i+1层神经元个数

## kaiming(He initialization)
Xavier 在tanh中表现很好，但在Relu函数中表现很差，故何凯明提出针对Relu的初始化方法

**基本思想**:在ReLU网络中，假定每一层有一半的神经元被激活，另一半为0，所以，要保持方差不变，只需要在 Xavier 的基础上再除以2。也就是说在方差推到过程中，式子左侧除以2

pytorch提供两种版本：
* `torch.nn.init.kaiming_uniform_(tensor, a=0, mode=‘fan_in’, nonlinearity=‘leaky_relu’)`均匀分布~U(−bound,bound),其中，bound的计算公式：![c][base64str3]
* `torch.nn.init.kaiming_normal_(tensor, a=0, mode=‘fan_in’, nonlinearity=‘leaky_relu’)`,正态分布~ N(0,std),其中,std的计算公式:![d][base64str4]
两个函数的参数:
* `a`:该层后面一层的激活函数中负的斜率(默认为ReLU，此时a=0)
* `mode`:‘fan_in’ (default) 或者 ‘fan_out’. 使用fan_in保持weights的方差在前向传播中不变；使用fan_out保持weights的方差在反向传播中不变

* **针对于Relu的激活函数，基本使用He initialization，pytorch也是使用kaiming 初始化卷积层参数的**

## 单位矩阵初始化
`torch.nn.eye_(tensor)`

将二维tensor初始化为单位矩阵(the identity matrix)
## 正交初始化
`torch.nn.init.orthogonal_(tensor, gain=1)`,使得`tensor`正交
## 稀疏初始化
`torch.nn.init.sparse_(tensor, sparsity, std=0.01)`,从正态分布N～（0,std）中进行稀疏化，使每一个column有一部分为0
`sparsity`:每一个column稀疏的比例，即为0的比例
